```python
import requests as req
import json
import urllib

import pandas as pd 
```

```python
key = '6497d1ddf07d485cbda142938181603' 
# coord = 'London'
coord = [[47.332,2.84],
[37.623317,-121.756267],
[-33.313662,-71.435548],
[-41.500531,173.786998],
[34.111158,-116.288460]]
```

```python
for i in coord:    
    url = "http://api.worldweatheronline.com/premium/v1/weather.ashx?query="+str(i[0])+","+str(i[1])+"&num_of_results=1&format=json&key="+key+"&cc=no&fx=no&includelocation=yes" 
    response = req.get(url)
    json_data = response.json()
    thing=json.dumps(json_data,indent=4, sort_keys=True) 
    print(thing)
```
